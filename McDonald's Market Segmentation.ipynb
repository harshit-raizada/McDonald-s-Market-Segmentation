#### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://homepage.boku.ac.at/leisch/MSA/datasets/mcdonalds.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
labels = ['Female', 'Male']
size = df['Gender'].value_counts()
explode = [0, 0.1]
colors = ['pink', 'yellow']

plt.rcParams['figure.figsize'] = (6, 6)
plt.pie(size, labels = labels, explode = explode, colors = colors, autopct = '%.2f%%', startangle = 90)
plt.title('Segmentation on Gender', fontsize = 15)
plt.legend()
plt.show()
```

#### We come to the conclusion that there are more female customers than male

```python
plt.rcParams['figure.figsize'] = (25, 8)
sns.countplot(x = df['Age'], palette = 'hsv')
plt.title('Distribution of Age', fontsize = 15)
plt.show()
```

#### McDonald's has customers from 50-60 years and 35-40 years

```python
df['Like'].value_counts()
```

#### Converting categorical value to numeric value

```python
df['Like'] = df['Like'].replace({'I hate it!-5': -5, 'I love it!+5': +5})
```

```python
sns.catplot(x = 'Like', y = 'Age', data = df, height = 4, aspect = 3, kind = 'bar')
plt.title('Likeliness w.r.t Age', fontsize = 15)
plt.show()
```

#### Converting categorical values to numeric values as it is easier to understand

```python
df['yummy'] = df['yummy'].replace({'No': 0, 'Yes': 1})
df['convenient'] = df['convenient'].replace({'No': 0, 'Yes': 1})
df['spicy'] = df['spicy'].replace({'No': 0, 'Yes': 1})
df['fattening'] = df['fattening'].replace({'No': 0, 'Yes': 1})
df['greasy'] = df['greasy'].replace({'No': 0, 'Yes': 1})
df['fast'] = df['fast'].replace({'No': 0, 'Yes': 1})
df['cheap'] = df['cheap'].replace({'No': 0, 'Yes': 1})
df['tasty'] = df['tasty'].replace({'No': 0, 'Yes': 1})
df['expensive'] = df['expensive'].replace({'No': 0, 'Yes': 1})
df['healthy'] = df['healthy'].replace({'No': 0, 'Yes': 1})
df['disgusting'] = df['disgusting'].replace({'No': 0, 'Yes': 1})
```

```python
df.head()
```

```python
df.corr()
```

```python
plt.figure(figsize = (12, 12))
sns.heatmap(df.corr(), annot = True)
plt.show()
```

```python
df['Gender'] = df['Gender'].replace({'Male': 1, 'Female': 0})
```

```python
sns.catplot(data = df, x = 'Like', y = 'Gender', kind = 'bar', aspect = 2, height = 5)
plt.title('Likeliness w.r.t Gender', fontsize = 15)
plt.show()
```

#### We can analyze from the catplot that gender does not play a vital role whether the customer will like it or not. It is also self explanatory.
